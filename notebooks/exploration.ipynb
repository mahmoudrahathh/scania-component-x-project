# Exploratory Data Analysis on Scania Component X Dataset

This notebook is used for performing exploratory data analysis (EDA) on the Scania Component X dataset. The dataset includes operational readouts, specifications, and time-to-event data. The goal of this EDA is to understand the data better and visualize key insights.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
```

```python
# Load the datasets
operational_readouts = pd.read_csv('../data/train_operational_readouts.csv')
specifications = pd.read_csv('../data/train_specifications.csv')
tte_data = pd.read_csv('../data/train_tte.csv')
```

```python
# Display the first few rows of each dataset
print("Operational Readouts:")
display(operational_readouts.head())

print("Specifications:")
display(specifications.head())

print("Time-to-Event Data:")
display(tte_data.head())
```

## Data Visualization

In this section, we will create visualizations to explore the relationships and distributions within the dataset.

```python
# Example visualization: Distribution of a specific operational metric
plt.figure(figsize=(10, 6))
sns.histplot(operational_readouts['some_metric'], bins=30, kde=True)
plt.title('Distribution of Some Metric')
plt.xlabel('Some Metric')
plt.ylabel('Frequency')
plt.show()
```
